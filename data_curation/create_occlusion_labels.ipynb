```python
# loop through all the occlusion labels from each object, create a new label file with the label from each object in 1 file.

base_data_dir = r'C:\Users\20195435\Documents\TUe\Pose Estimation\data\LINEMOD_updated\LINEMOD'
occlusion_dir = r'C:\Users\20195435\Documents\TUe\Pose Estimation\data\LINEMOD_updated\LINEMOD\occlusion'


objects = ["ape", "can", "cat", "driller", "duck", "eggbox", "glue", "holepuncher"]

objects_dict = {}

label_file_list = []

for object in objects:
    object_dir = os.path.join(base_data_dir, object)
    object_occlusion_dir = os.path.join(base_data_dir, object, 'labels_occlusion')

    
    object_occlusion_label_files = os.listdir(object_occlusion_dir)
    # add all labels to a dict with the file name as key
    labels = {}
    for label_file in object_occlusion_label_files:
        if label_file not in label_file_list:
            label_file_list.append(label_file)
        with open(os.path.join(object_occlusion_dir, label_file), 'r') as f:
            labels[label_file] = f.readlines()

    objects_dict[object] = labels


for label_file in label_file_list:
    with open(os.path.join(occlusion_dir, label_file), 'w') as f:
        for object in objects:

            if label_file in objects_dict[object]:
                for line in objects_dict[object][label_file]:
                    f.write(line)
```

```python

labels_folder = r"C:\Users\20195435\Documents\TUe\Pose Estimation\data\LINEMOD_updated\LINEMOD\occlusion\labels_old"
new_labels_folder = r"C:\Users\20195435\Documents\TUe\Pose Estimation\data\LINEMOD_updated\LINEMOD\occlusion\labels"


for label_file in os.listdir(labels_folder):
    with open(os.path.join(labels_folder, label_file), 'r') as f:
        lines = f.readlines()
    with open(os.path.join(new_labels_folder, label_file), 'w') as f:
        for line in lines:
            label = line.split()
            label[0] = str(mapping[int(float(label[0]))])
            f.write(' '.join(label) + '\n')
```
